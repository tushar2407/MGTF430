```python
import pandas as pd
import numpy as np
import numpy as np
from sklearn.mixture import GaussianMixture
```

```python
data = pd.read_csv('CRSP_hitorical_stock_daily_data.csv')
data.head()
```

```python
data.shape
```

```python
data.groupby('PERMNO').size()
```

```python
data_sub = data[data['PERMNO'] == 10026]
data_sub
```

```python
# get all the numerical columns
X = data_sub.select_dtypes(include=[np.number])
X = X.dropna(axis=1)
X
```

```python
X.columns
```

# Gaussian Mixture Model (GMM)

```python

gm = GaussianMixture(n_components=2, random_state=0).fit(X)
gm.means_
```

```python
gm.means_.shape
```

```python
X_reduced = gm.predict_proba(X.to_numpy())
X_reduced
```

# LDA (LSI)

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
X_reduced_pca = pca.fit_transform(X)
X_reduced_pca
```

# KMeans

```python
from sklearn.cluster import KMeans
```

```python
X_reduced_kmeans = KMeans(n_clusters=2, random_state=0, n_init="auto").fit_transform(X)
X_reduced_kmeans
```

# LDA

```python
from sklearn.decomposition import LatentDirichletAllocation as LDA
```

```python
X_reduced_lda = LDA().fit_transform(X)
X_reduced_lda
```
